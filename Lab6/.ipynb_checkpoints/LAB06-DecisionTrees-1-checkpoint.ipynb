**Read the data**

```python
import os
import pandas as pd
#data_path = os.path.join("datasets", "divorce.csv")
df = pd.read_csv("divorce.csv")
df.head(10)
```

**Tasks 1:** Randomly split the dataset into train_X, train_Y, test_X, and test_Y with 80% train and 20% test. [10 points]

```python
# TODO
from sklearn.model_selection import train_test_split

X = df.loc[:, df.columns != 'Class']
y = df['Class']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

**Task 2:** Train a DecisionTreeClassifier based on train_X and train_Y. Report the performance on test data. [10 points]

```python
from sklearn import tree
from sklearn import metrics

# TODO
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)

print("%0.3f" % clf.score(x_train, y_train))
```

**Task 3:** Set the max_depth to 3. Retrain the model and report the performance on test data. Please briefly explain why performances change. [10 points]

```python
from sklearn import tree
from sklearn import metrics

# TODO
clf = tree.DecisionTreeClassifier(max_depth = 3)
clf = clf.fit(x_train, y_train)

print("%0.3f" % clf.score(x_train, y_train))

# max_depth allows the tree to go deeper. The higher the max_depth, then the more complex model it will become.
# However, for the case when the performance changes, it is 
# because the model overfits for bigger depth values. The model able to predict all of the train data, however, 
# fails to generalize the finding for new data.
```
