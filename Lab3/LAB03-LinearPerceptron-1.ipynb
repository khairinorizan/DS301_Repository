## 1. Get and visualize the data to gain insights

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##data_path = os.path.join("datasets", "german_credit_data.csv")
data = pd.read_csv("german_credit_data.csv")
```

```python
import matplotlib.pyplot as plt
data['Risk'].hist()
plt.xlabel('Risk')
plt.ylabel('Count')
```

## 2. Separate the data into features (x) and label (y) such that

- y is a binary target, indicating whether risk is "good"
- x contains the following numerical columns: Age, Job, Credit amount, Duration.

```python
attributes = ['Age', 'Job', 'Credit amount', 'Duration']

x = data[attributes]

y = (data['Risk'] == 'good')
```

** Task 1:** Use scikit-learn to train your model with LinearPerceptron. [10 points]

```python
from sklearn.datasets import load_digits
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split

responses = x.to_numpy()
predictors = y.to_numpy()

X_train, X_test, Y_Train, Y_Test = train_test_split(responses, predictors, test_size = 0.2, random_state = 20)

perceptronModel = Perceptron(random_state=1)
perceptronModel.fit(responses, predictors)
```

**Task 2:** Use matplotlib to visualize the data. Use 'Credit amount' as x-axis and 'Duration' as y-axis and assign different colors for data with different labels. Also plot your model decision boundary. [10 points]

```python
import matplotlib.cm as cm

plt.figure(figsize=(12,6))

xValues = data['Credit amount']
yValues = data['Duration']

yValues

plt.title('Credit Amount vs Duration')
plt.xlabel('Credit amount')
plt.ylabel('Duration')

colors = {True: 'green', False: 'red'}

plt.scatter(xValues, yValues, c=y.map(colors))
plt.plot(xValues, yValues, color = 'blue', linewidth=0.1, animated=True)
plt.show()
```

**Task 3:** Report the training accuracy. [10 points]

```python
print("%0.3f" % perceptronModel.score(responses, predictors))
```
